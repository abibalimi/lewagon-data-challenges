```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
```

```python
import requests

url = "http://iex.lewagon.com/stable/stock/aapl/chart/3m"
api_data = requests.get(url).json()
```

```python
api_data
```

```python
apple_df = pd.DataFrame(api_data)
apple_df
```

```python
apple_df.info()
```

```python
apple_df['date'] = pd.to_datetime(apple_df['date'], format="%Y-%m-%d")
```

```python
apple_df = apple_df.set_index('date')
apple_df
```

```python
plot = apple_df['close'].plot()
```

```python
apple_df[['open','close','high','low']].plot(figsize=(12, 4))
```
