# Subplotting

Sometimes we need more than one plot, we might want to compare 3 similar graphs or show an evolution.
To do so we can create subplots in our `figure` object to create a **grid** of plots.

<img src="https://matplotlib.org/3.1.1/_images/sphx_glr_axes_margins_001.png" alt="subplot" width=400>

### Side by side plot

Change the following code to show the two barplots **side by side**.
To do this use the [plt.subplots()](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplots.html) function.

<details>
  <summary>ðŸ’¡Hint</summary>
  <p>You will need to use axes.</p>
</details>

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

wallet_2015 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234,74,530]}
wallet_2020 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234*4,74*7,530*1.3]}

# CHANGE THE FOLLOWING CODE
ax1.bar(wallet_2015["currency"], wallet_2015["value_in_euro"])
ax2.bar(wallet_2020["currency"], wallet_2020["value_in_euro"])
plt.show()
```

```python
bitcoin = [3,3,4,5,5,6,11,12,13,14,16,14,18,15,19,18,19,15,10,8,9,9,9,12,13]
ethereum = [1,1,1,2,2,3,3,4,4,4,4,8,9,12,10,9,14,15,10,8,6,5,5,8,7]
euro = [7,7,7,7.6,7,7.8,7,7,7,7,7.6,7,7,7,7,6.8,6.9,7,7,7,7,7,7.2,7.2,7.3]

# CHANGE THE FOLLOWING CODE
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax1.bar(wallet_2015["currency"], wallet_2015["value_in_euro"])

ax2 = fig.add_subplot(222)
ax2.bar(wallet_2020["currency"], wallet_2020["value_in_euro"])

ax3 = fig.add_subplot(212)
ax3.plot(bitcoin)
ax3.plot(ethereum)
ax3.plot(euro)
plt.show()
```

```python
# YOUR SUBPLOTS HERE
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(3, 3)
f3_ax1 = fig3.add_subplot(gs[:2, :2])
f3_ax1.set_title('gs[:2, :2]')
f3_ax2 = fig3.add_subplot(gs[:1, -1])
f3_ax2.set_title('gs[1, 3]')
f3_ax3 = fig3.add_subplot(gs[1:2, -1])
f3_ax3.set_title('gs[2, 3]')
f3_ax4 = fig3.add_subplot(gs[-1, -1])
f3_ax4.set_title('gs[3, 3]')
f3_ax5 = fig3.add_subplot(gs[-1, 1:2])
f3_ax5.set_title('gs[3, 2]')
f3_ax6 = fig3.add_subplot(gs[-1, :1])
f3_ax6.set_title('gs[3, 1]');
```

```python
# YOUR SUBPLOTS HERE
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(4, 3)
f3_ax1 = fig3.add_subplot(gs[0, 0])
f3_ax1.set_title('gs[0, 0]')
f3_ax2 = fig3.add_subplot(gs[0, 1:])
f3_ax2.set_title('gs[0, 1:]')
f3_ax3 = fig3.add_subplot(gs[1, :])
f3_ax3.set_title('gs[1, :]')
f3_ax4 = fig3.add_subplot(gs[2:, 0])
f3_ax4.set_title('gs[2:, 0]')
f3_ax5 = fig3.add_subplot(gs[2:, 1:])
f3_ax5.set_title('gs[:2, 1:]');
```
