The dataset is made up of email that are classified as ham [0] or spam[1]. You need to clean the dataset before training a prediction model.

```python
X = df.drop(columns='spam').copy()
y = df['spam']
```

```python
X
```

```python
X['clean_text'] = X['text']
X.head()
```

## Remove Punctuation

👇 Create a function to remove the punctuation. Apply it to the entire data and add the output as a new column in the dataframe called `clean_text`

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
X.clean_text = X.clean_text.apply(remove_punctuation)
X
```

```python
X.info()
```

## Lower Case

👇 Create a function to lower case the text. Apply it to `clean_text`

```python
def apply_lower_case(text):
    return text.lower()
```

```python
X.clean_text = X.clean_text.apply(apply_lower_case)
X
```

## Remove Numbers

👇 Create a function to remove numbers from the text. Apply it to `clean_text`

```python
def remove_numbers(text):
    return ''.join(word for word in text if not word.isdigit())
```

```python
X.clean_text = X.clean_text.apply(remove_numbers)
```

```python
X
```

## Remove StopWords

👇 Create a function to remove stopwords from the text. Apply it to `clean_text`.

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english')) 
```

```python
def remove_stopwords(text):
    word_tokens = word_tokenize(text)
    text = [word for word in word_tokens if not word in stop_words]
    return text
```

```python
X.clean_text = X.clean_text.apply(remove_stopwords)
X
```

## Lemmatize

👇 Create a function to lemmatize the text. Make sure the output is a single string, not a list of words. Apply it to `clean_text`.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
X.clean_text[0]
```

```python
def lemmatize_them(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text])
```

```python
#ct = X.clean_text.apply(lemmatize_them)
X.clean_text = X.clean_text.apply(lemmatize_them)
X
```

```python
X.clean_text[0]
```

## Bag-of-words Modelling

👇 Vectorize the `clean_text` to a Bag-of-Words representation with a default CountVectorizer . Save as `X_bow`.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
```

```python
X_bow = vectorizer.fit_transform(X.clean_text)

X_bow.toarray()
```

```python
pd.DataFrame(X_bow.toarray(),columns = vectorizer.get_feature_names())
```

👇 Cross-validate a MultinomialNB model with the Bag-of-words. Score the model's accuracy.

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB

nb_model = MultinomialNB()
```

```python
nb_model.fit(X_bow, y)
nb_model.score(X_bow, y)
```

```python
cross_val_score(nb_model, X_bow, y, scoring='accuracy', cv=5).mean()
```
